```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
from statsmodels.tsa.arima.model import ARIMA
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Conv1D, MaxPooling1D, GRU
from tensorflow.keras.optimizers import Nadam

# Load and preprocess data
data = pd.read_csv('663221_52.49_11.02_2019.csv', skiprows=1)
df = data
df.columns = df.iloc[0]
df = df.iloc[1:]
df = df.iloc[0:,:10]
df = df.reset_index(drop=True).reset_index()
df = df.rename(columns={'index':'Index'})
df = df.set_index('Index')

data_selected = df[['Temperature', 'Relative Humidity', 'Pressure', 'Wind Speed', 'GHI']].apply(pd.to_numeric, errors='coerce')

# Plotting scatter plots for each of the selected columns
plt.figure(figsize=(12, 8))
for i, column in enumerate(data_selected.columns, 1):
    plt.subplot(len(data_selected.columns), 1, i)
    plt.scatter(data_selected.index, data_selected[column], s=10, alpha=0.5)
    plt.title(f'Scatter Plot for {column}')
    plt.xlabel('Index')
    plt.ylabel(column)

plt.tight_layout()
plt.show()

# Normalize the data
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns, index=df.index)

# Prepare data for time series models
X = df_scaled.drop('GHI', axis=1)
y = df_scaled['GHI']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
X_val, X_test, y_val, y_test = train_test_split(X_test, y_test, test_size=0.5, shuffle=False)

# Model Development

def train_arima(y_train):
    model = ARIMA(y_train, order=(1,1,1))
    return model.fit()

def create_lstm_model(input_shape):
    model = Sequential([
        LSTM(50, return_sequences=True, input_shape=input_shape),
        LSTM(50),
        Dense(1)
    ])
    model.compile(optimizer=Nadam(), loss='mse')
    return model

def create_cnn_lstm_tf_model(input_shape):
    model = Sequential([
        Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=input_shape),
        MaxPooling1D(pool_size=2),
        LSTM(50, return_sequences=True),
        LSTM(50),
        Dense(1)
    ])
    model.compile(optimizer=Nadam(), loss='mse')
    return model

def create_gru_model(input_shape):
    model = Sequential([
        GRU(50, return_sequences=True, input_shape=input_shape),
        GRU(50),
        Dense(1)
    ])
    model.compile(optimizer=Nadam(), loss='mse')
    return model

def create_arima_lstm_hybrid(y_train, input_shape):
    arima_model = train_arima(y_train)
    arima_residuals = y_train - arima_model.fittedvalues
    
    lstm_model = create_lstm_model(input_shape)
    lstm_model.fit(X_train, arima_residuals, epochs=10, batch_size=32, validation_data=(X_val, y_val))
    
    return arima_model, lstm_model

# Model Training and Optimization

arima_model = train_arima(y_train)

lstm_model = create_lstm_model((X_train.shape[1], 1))
lstm_model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_val, y_val))

cnn_lstm_tf_model = create_cnn_lstm_tf_model((X_train.shape[1], 1))
cnn_lstm_tf_model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_val, y_val))

gru_model = create_gru_model((X_train.shape[1], 1))
gru_model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_val, y_val))

arima_lstm_model = create_arima_lstm_hybrid(y_train, (X_train.shape[1], 1))

# Evaluation and Comparison

def evaluate_model(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mape = mean_absolute_percentage_error(y_true, y_pred)
    return mae, rmse, mape

arima_pred = arima_model.forecast(steps=len(y_test))
arima_mae, arima_rmse, arima_mape = evaluate_model(y_test, arima_pred)

lstm_pred = lstm_model.predict(X_test)
lstm_mae, lstm_rmse, lstm_mape = evaluate_model(y_test, lstm_pred)

cnn_lstm_tf_pred = cnn_lstm_tf_model.predict(X_test)
cnn_lstm_tf_mae, cnn_lstm_tf_rmse, cnn_lstm_tf_mape = evaluate_model(y_test, cnn_lstm_tf_pred)

gru_pred = gru_model.predict(X_test)
gru_mae, gru_rmse, gru_mape = evaluate_model(y_test, gru_pred)

arima_forecast = arima_lstm_model[0].forecast(steps=len(y_test))
lstm_residuals = arima_lstm_model[1].predict(X_test).flatten()
hybrid_pred = arima_forecast + lstm_residuals
hybrid_mae, hybrid_rmse, hybrid_mape = evaluate_model(y_test, hybrid_pred)


#Visualization and Analysis
plt.figure(figsize=(12, 6))
plt.plot(y_test.index, y_test, label='Actual')
plt.plot(y_test.index, arima_pred, label='ARIMA')
plt.plot(y_test.index, lstm_pred, label='LSTM')
plt.plot(y_test.index, cnn_lstm_tf_pred, label='CNN-LSTM-TF')
plt.plot(y_test.index, gru_pred, label='GRU')
plt.plot(y_test.index, hybrid_pred, label='ARIMA-LSTM Hybrid')
plt.title('Solar Power Production: Actual vs Predicted')
plt.xlabel('Date')
plt.ylabel('Normalized GHI')
plt.legend()
plt.show()

# Print evaluation metrics
print("ARIMA - MAE: {:.4f}, RMSE: {:.4f}, MAPE: {:.4f}".format(arima_mae, arima_rmse, arima_mape))
print("LSTM - MAE: {:.4f}, RMSE: {:.4f}, MAPE: {:.4f}".format(lstm_mae, lstm_rmse, lstm_mape))
print("CNN-LSTM-TF - MAE: {:.4f}, RMSE: {:.4f}, MAPE: {:.4f}".format(cnn_lstm_tf_mae, cnn_lstm_tf_rmse, cnn_lstm_tf_mape))
print("GRU - MAE: {:.4f}, RMSE: {:.4f}, MAPE: {:.4f}".format(gru_mae, gru_rmse, gru_mape))
print("ARIMA-LSTM Hybrid - MAE: {:.4f}, RMSE: {:.4f}, MAPE: {:.4f}".format(hybrid_mae, hybrid_rmse, hybrid_mape))

# Analyze impact of weather parameters
correlation_matrix = df[['Temperature', 'Relative Humidity', 'Pressure', 'Wind Speed', 'GHI']].corr()
plt.figure(figsize=(10, 8))
plt.imshow(correlation_matrix, cmap='coolwarm')
plt.colorbar()
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.title('Correlation between Weather Parameters and Solar Power Output')
plt.show()
```
